```python
#Lab1: Write a Pandas program to split the following data frame into groups based on Class and count the number of students 
#in that particular class.Also generate a bar chart based on the result and explain the conclusion.

import pandas as pd
import matplotlib.pyplot as plt

student_data = pd.DataFrame({
'school_code': ['s001','s002','s003','s001','s002','s004'],
'class': ['V', 'VI', 'VI', 'VI', 'V', 'VI'],
'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill','David Parkes'],
'age': [12, 12, 13, 13, 14, 12],
'height': [173, 192, 186, 167, 151, 159],
'weight': [35, 32, 33, 30, 31, 32],
'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
)

print("Original Dataframe: \n\n",student_data)

class_counts = student_data.groupby('class').size()       #Grouping by 'class' and counting the number of students in each class

plt.figure(figsize=(8, 5))
class_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Class')
plt.xticks(rotation=0)
plt.show()
print('''\nThis bar chart is showing the number of students in each class. From the counts, we can conclude which class has the most or the fewest students.
For the given data:
Class VI has 4 students.
Class V has 2 students.
Thus, Class VI is the most populated class in this dataset, indicating a higher concentration of students compared to Class V.''')
```

```python
#Lab2: Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school.
#Also generate a horizontal bar chart based on the result and explain the conclusion.

student_data = pd.DataFrame({
'school_code': ['s001','s002','s003','s001','s002','s004'],
'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill',
'David Parkes'],
'age': [12, 12, 13, 13, 14, 12],
'height': [173, 192, 186, 167, 151, 159],
'weight': [35, 32, 33, 30, 31, 32],
'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
)

print("Original Dataframe:\n\n",student_data)

age_stats = student_data.groupby('school_code')['age'].agg(['mean', 'min', 'max'])   #Grouping by 'school_code' and calculating mean, min, and max age
student_data.set_index('school_code', inplace=True)
print("\nMean, min and max value of age for each value of the school: \n\n",age_stats)
print("\n\n")
fig, ax = plt.subplots(figsize=(10, 6))
age_stats.plot(kind='barh', color=['skyblue', 'lightgreen', 'salmon'], ax=ax)
ax.set_xlabel('Age')
ax.set_ylabel('School Code')
ax.set_title('Mean, Min, and Max Age by School Code')
plt.legend(title='Age Statistics')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
print('''\nThis horizontal bar chart is showing the mean, min, and max ages for each school code. By examining the chart, we can make the following observations:
1) Compare the central tendency (mean) and the range (min to max) of ages across different schools.
2)Identify if there is significant variation in the age distribution within or between schools.

For this data, we will observe:
School Code s001 has a mean and max age of 13, with a min of 12.
School Code s002 has a mean age of 13, with a min of 12 and a max of 14.
School Code s003 has a mean, min, and max age of 13.
School Code s004 has a mean, min, and max age of 12.
These statistics can help in understanding the age distribution across different schools.''')
```

```python
#Lab3: Write a Pandas program to split a dataset, group by one column and get mean, min, and max values by group. 
#Using the following dataset find the mean, min, and max values of purchase amount (purch_amt) group by
#customer id (customer_id).Also generate a line chart based on the result and explain the conclusion.

orders_data = pd.DataFrame({
'ord_no':[70001,70009,70002,70004,70007,70005,70008,70010,70003,70012,70011,70013],
'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45,75.29,3045.6],
'ord_date':['2012-10-05','2012-09-10','2012-10-05','2012-08-17','2012-09-10','2012-07-27','2012-09-10','2012-10-10','2012-10-10','2012-06-27','2012-08-17','2012-04-25'],
'customer_id':[3005,3001,3002,3009,3005,3007,3002,3004,3009,3008,3003,3002],
'salesman_id': [5002,5005,5001,5003,5002,5001,5001,5006,5003,5002,5007,5001]})

print("Original Order's Dataframe:\n\n",orders_data)
purch_stats = orders_data.groupby('customer_id')['purch_amt'].agg(['mean', 'min', 'max']).reset_index()   # Grouping by 'customer_id' and calculating mean, min, and max purchase amount

print("\n",purch_stats)


plt.figure(figsize=(10, 6))
plt.plot(purch_stats['customer_id'], purch_stats['mean'], marker='o', label='Mean Purchase Amount', color='blue')
plt.plot(purch_stats['customer_id'], purch_stats['min'], marker='o', label='Min Purchase Amount', color='green')
plt.plot(purch_stats['customer_id'], purch_stats['max'], marker='o', label='Max Purchase Amount', color='red')
plt.xlabel('Customer ID')
plt.ylabel('Purchase Amount')
plt.title('Mean, Min, and Max Purchase Amount by Customer ID')
plt.legend(title='Purchase Amount Statistics')
plt.grid(True)
plt.show()
print('''\nThis line chart is showing how the mean, min, and max purchase amounts vary across different customers. 
By analyzing the chart:
Customer ID 1 has the highest variability in purchase amounts, with a wide range between the min and max values.
Customer ID 2 has a lower max amount compared to Customer ID 1 but also shows variability.
Customer ID 3 has a single data point, so mean, min, and max are the same.
Customer ID 4 has similar min and max values, indicating less variability in purchases.
Overall, the line chart is visually representing the distribution and range of purchase amounts for each customer, helping us understand purchasing behavior better. ''')
```

```python
#Lab4: Write a Pandas program to split the following data frame into groupsand calculate monthly purchase amount.
#Also generate a bar chart based on the result and explain the conclusion.

df = pd.DataFrame({
'ord_no': [70001, 70009, 70002, 70004, 70007, 70005, 70008, 70010, 70003, 70012, 70011, 70013],
    'purch_amt': [150.5, 270.65, 65.26, 110.5, 948.5, 2400.6, 5760, 1983.43, 2480.4, 250.45, 75.29, 3045.6],
    'ord_date': ['05-10-2012', '09-10-2012', '05-10-2012', '08-17-2012', '10-09-2012', '07-27-2012', '10-09-2012', '10-10-2012', '10-10-2012', '06-17-2012', '07-08-2012', '04-25-2012'],
    'customer_id': [3001, 3001, 3005, 3001, 3005, 3001, 3005, 3001, 3005, 3001, 3005, 3005],
    'salesman_id': [5002, 5005, 5001, 5003, 5002, 5001, 5001, 5006, 5003, 5002, 5007, 5001]
})

# Convert 'ord_date' to datetime format
df['ord_date'] = pd.to_datetime(df['ord_date'], format='%m-%d-%Y')

# Extract year and month
df['year_month'] = df['ord_date'].dt.to_period('M')

# Group by 'year_month' and sum the purchase amounts
monthly_purchases = df.groupby('year_month')['purch_amt'].sum().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(monthly_purchases['year_month'].astype(str), monthly_purchases['purch_amt'], color='skyblue')
plt.xlabel('Month-Year')
plt.ylabel('Total Purchase Amount')
plt.title('Monthly Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print('''\nThis bar chart is displaying the total purchase amount for each month.
By examining the chart:
1) We can identify which months had the highest and lowest purchase amounts.
2) WE can observe trends over the months, such as any increase or decrease in spending.

For the given sample data:
January 2024: Total purchases are the sum of all amounts in January.
February 2024: Total purchases are the sum of all amounts in February.
March 2024: Total purchases are the sum of all amounts in March.
This visualization helps in understanding purchasing behavior trends over time and identifying any seasonal patterns or anomalies in the data. ''')
```

```python
#Scenario:Analyzing E-commerce Customer Behavior
#Suppose you work for an e-commerce company, and you have a dataset containing information about customer behavior on your website.
#The dataset includes data on customer visits, product views, purchases, and more. You want to perform data analysis
#like Calculate the total number of unique customers,Calculate the total number of visits,Calculate the average number of products viewed per visit,
#Identify the most viewed product,Calculate the total revenue, Visualize the daily visit trend using Pandas to gain insights into customer behavior.
#Further, you need to get some inference out of the chart.


import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'customer_id': [1, 1, 2, 2, 3, 3, 1, 4, 5, 5],
    'visit_date': ['2024-08-01', '2024-08-01', '2024-08-02', '2024-08-03', '2024-08-02', '2024-08-04', '2024-08-05', '2024-08-05', '2024-08-05', '2024-08-06'],
    'product_id': [101, 102, 101, 103, 104, 105, 101, 102, 103, 104],
    'purchase_amount': [20, 30, 15, 25, 35, 20, 30, 40, 50, 60]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Convert 'visit_date' to datetime
df['visit_date'] = pd.to_datetime(df['visit_date'])

# 1. Calculate the total number of unique customers
total_unique_customers = df['customer_id'].nunique()

# 2. Calculate the total number of visits
total_visits = df.groupby('customer_id')['visit_date'].count().sum()

# 3. Calculate the average number of products viewed per visit
products_per_visit = df.groupby(['customer_id', 'visit_date'])['product_id'].count()
avg_products_per_visit = products_per_visit.mean()

# 4. Identify the most viewed product
most_viewed_product = df['product_id'].value_counts().idxmax()

# 5. Calculate the total revenue
total_revenue = df['purchase_amount'].sum()

# 6. Visualize the daily visit trend
daily_visits = df.groupby('visit_date')['customer_id'].count()

plt.figure(figsize=(10, 6))
daily_visits.plot(kind='line', marker='o', color='blue')
plt.xlabel('Date')
plt.ylabel('Number of Visits')
plt.title('Daily Visit Trend')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Output results
print("Total Unique Customers:", total_unique_customers)
print("Total Visits:", total_visits)
print("Average Number of Products Viewed Per Visit:", avg_products_per_visit)
print("Most Viewed Product ID:", most_viewed_product)
print("Total Revenue:", total_revenue)
print('''\nConclusion:
Unique Customers: Indicates the number of distinct users, showing the breadth of your customer base.
Total Visits: Reflects overall traffic. If high compared to unique customers, it suggests frequent visits by the same users.
Average Products Viewed Per Visit: Shows customer engagement. A higher average means users are exploring more products per visit.
Most Viewed Product: Highlights popular items and helps focus marketing efforts.
Total Revenue: Indicates overall sales performance and financial health.

Daily Visit Trend Chart:
Trends: Identifies peak and low periods, helping link traffic changes to marketing activities or seasonal patterns.
Patterns: Reveals consistent behavior over time, like increased visits on specific days or seasons.

Overall, the data helps understand customer engagement, popular products, and financial performance, guiding marketing and site improvements. ''')
```
