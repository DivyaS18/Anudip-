```python
#Lab1: Analyze the relationship between the size of houses (measured in square footage) and their selling prices in a particular neighborhood.
#You have collected data on various houses in that neighborhood.Create a scatter plot using the below data and share your conclusion/analysis.

import numpy as np
import matplotlib.pyplot as plt

square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800,3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])
plt.figure(figsize=(10, 6))
plt.scatter(square_footage, selling_prices, color='blue', marker='o')
plt.title('House Size vs. Selling Price')
plt.xlabel('Square Footage')
plt.ylabel('Selling Price (in thousands)')
plt.grid(True)
plt.show()
```

```python
#Lab2: Create a pie chart to visualize the distribution of your monthly income by source. You have collected data on the various sources
#of your income, such as salary, freelance work, investments, and rental income. Share your conclusion/analysis.

income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

plt.figure(figsize=(6, 6))
plt.pie(monthly_income, labels=income_sources, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Monthly Income Distribution by Source')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = {
    'SalesDate': pd.date_range(start='2023-01-01', periods=12, freq='ME'),
    'ProductName': np.random.choice(['Product A', 'Product B', 'Product C', 'Product D'], 12),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 12),
    'SalesQuantity': np.random.randint(10, 100, 12),
    'Price': np.random.randint(20, 200, 12)
}

df = pd.DataFrame(data)

# Bar Chart: Total Sales Quantity per Product
product_sales = df.groupby('ProductName')['SalesQuantity'].sum()
plt.figure(figsize=(5, 5))
plt.bar(product_sales.index, product_sales.values, color='skyblue')
plt.title('Total Sales Quantity per Product')
plt.xlabel('Product Name')
plt.ylabel('Total Sales Quantity')
plt.grid(axis='y')

# Pie Chart: Distribution of Total Sales by Region
region_sales = df.groupby('Region')['SalesQuantity'].sum()
plt.figure(figsize=(5, 5))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Total Sales Distribution by Region')

# Pie Chart: Distribution of Total Sales by Price Range
price_ranges = pd.cut(df['Price'], bins=[0, 50, 100, 150, 200], labels=['20-50', '51-100', '101-150', '151+'])
price_range_sales = df.groupby(price_ranges, observed=False)['SalesQuantity'].sum()
plt.figure(figsize=(5, 5))
plt.pie(price_range_sales, labels=price_range_sales.index, autopct='%1.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Total Sales Distribution by Price Range')

# Line Chart: Trend of Average Prices Over Time
df['Month'] = df['SalesDate'].dt.to_period('M')
monthly_avg_price = df.groupby('Month')['Price'].mean()
plt.figure(figsize=(5, 5))
plt.plot(monthly_avg_price.index.astype(str), monthly_avg_price.values, marker='o', color='green')
plt.title('Average Prices Over Time')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```
