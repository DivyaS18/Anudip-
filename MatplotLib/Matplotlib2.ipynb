```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,80,90]
y=[1,2,3,4,5,6,7,8,9]
plt.subplot(2,2,1)
plt.plot(x,y,color='g')

plt.subplot(2,2,2)
plt.pie([1],colors='r')

plt.subplot(2,2,3)
a=[20,40,60]
plt.pie(a)

plt.subplot(2,2,4)
a1=["r","a","h","u"]
a2=[20,40,60,80]
plt.bar(a1,a2)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.plot(x,y)
plt.title("Plot with general text")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.text(2,8, 'Language', fontsize=10, color='white', bbox={"facecolor":"red"})
plt.annotate(r'Java', xy=(4,7), xytext=(2,10), arrowprops=dict(facecolor='yellow', shrink=0.10), fontsize=12, color='red')
plt.legend(["Java"], loc=5, facecolor="yellow", edgecolor="red", framealpha=0.9, shadow=True)
plt.show()
```

```python
#Matplotlib provides several functions to customize the appearance of the axes in a plot. These includes
#setting the ticks and limits for both the x-axis and y-axis. Here is how you can use xticks, yticks, xlim, and ylim in Matplotlib, along with some practical examples.
#xticks and yticks:These functions are used to get or set the current tick locations and labels of the x-axis and y-axis, respectively.
#xlim and ylim: These functions are used to get or set the current x-axis and y-axis limits, respectively.
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)         #creating plot
plt.xlim(0,10)
plt.ylim(0,12)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)

plt.xticks([1,2,3,4,5], ['One', 'Two', 'Three', 'Four', 'Five'])
plt.yticks([2,4,6,8,10,12], ['Tow','Four', 'Six', 'Eight', 'Ten', 'Twelve'])

plt.title('Custom Ticks Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
