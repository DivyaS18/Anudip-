```python
#1 Calculate the total revenue generated by 2 product categories in a store.
#Input: category1_revenue = np.array([500,600,700,550])
#category2_revenue = np.array([450,700,800,600])
#Output: Total Revenue: [950,1300 1500 1150]
import numpy as np

category1_revenue = np.array([500,600,700,550])
category2_revenue = np.array([450,700,800,600])
total_revenue=category1_revenue+category2_revenue
print("Total revenue:",total_revenue)
```

```python
#2 Calculate the profit made by a company
#Input: revenue=np.array([10000,12000,11000,10500])
#expenses=np.array([4000,5000,4500,4800])
#output: profit:[6000,7000,6500,5700]

revenue=np.array([10000,12000,11000,10500])
expenses=np.array([4000,5000,4500,4800])
profit=revenue-expenses
print("Profit: ",profit)
```

```python
#3Determine which products in a store are out of stock(quantity 0).
#input:inventory=np.array[10,0,5,0,20,0])
#output:out of stock products:[0 0 0]

inventory=np.array([10,0,5,0,20,0])
out_of_stock_products = inventory[inventory == 0]
print("Out of stock products:", out_of_stock_products)
```

```python
#4Calculate the total cost of items in a shopping cart,considering the quantity and price per item.
#input:quantity=np.array([2,3,4,1])
#price_per_item=np.array([10.0,5.0,8.0,12.0])
#output: the cost of items:[20,15,32,12]

quantity=np.array([2,3,4,1])
price_per_item=np.array([10.0,5.0,8.0,12.0])
total_cost=quantity*price_per_item
print("The cost of items: ",total_cost)
```
